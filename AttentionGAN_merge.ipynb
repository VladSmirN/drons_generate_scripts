```python
def binarize(img):

  #initialize threshold
  thresh=255

  #convert image to greyscale
  img=img.convert('L')

  width,height=img.size


  #traverse through pixels
  for x in range(width):
    for y in range(height):
      r = 25
      x0 = x-width/2
      y0 = y-height/2

      pixel = img.getpixel((x,y))
      if pixel < thresh:
        img.putpixel((x,y),255)
      else:
        img.putpixel((x,y),0)

      if  x0*x0+y0*y0>r*r :
        img.putpixel((x,y),0)

  return img

for i,row in fake_B_df.iterrows():
    image_generate = Image.open(row["path_image"])
    image_mask = Image.open(row["mask_path"])
    image_mask =  binarize(image_mask)

    mask_im = Image.new("L", image_generate.size, 0)
    draw = ImageDraw.Draw(mask_im)
    draw.ellipse((30, 30, 70, 70), fill=255)
    mask_im_blur = mask_im.filter(ImageFilter.GaussianBlur(3))



    image_original = Image.open(row["original_path"])
    image_original.paste(image_generate,(row["x0"],row["y0"]),image_mask)
    if row['id_image'] in no_anamaly_merge:
        Image.open(row["original_path"]).save(os.path.join(SAVE_DIR,f"{row['id_image']}.jpg"))
        # image_original.crop((row["x0"], row["y0"], row["x0"]+100, row["y0"]+100)).save(os.path.join(SAVE_DIR,f"{row['id_image']}.jpg"))

    # image_original.save(os.path.join(SAVE_DIR,f"{row['id_image']}.jpg"))

    # image_original = Image.open(row["original_path"])
    # image_original.paste(image_generate,(row["x0"],row["y0"]))
    # image_original.save(os.path.join(SAVE_DIR,f"{row['id_image']}_2.jpg"))

    # image_mask.save(os.path.join(SAVE_DIR,f"{row['id_image']}_m.jpg"))
    # image_original.show( )
    # break
    print(row["id"])
```
